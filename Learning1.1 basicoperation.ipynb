* ### Screen resolution
#### For example, the screen resolution is 1024 Ã— 768, that is, there are 1024 pixels in the horizontal direction and 768 pixels in the vertical direction of the device screen. The pixel size has no fixed size.
#### For example, for two screens with the same size, the resolution can be different. The screen with higher resolution has more pixels, so the screen can display more detailed images, and the area of a single pixel block is smaller. On the screen with low resolution, there are fewer pixels, and the area of a single block is larger, so the screen display less detail.

* ### Image resolution
#### For example, a image resolution is 500x200, that is, when the picture is magnified 1:1 on the screen, there are 500 pixels in the horizontal direction and 200 pixels in the vertical direction.
#### Resolution=pixel number in the horizontal direction of the image* pixel number in the vertical direction of the image
